```python
## First lets handle Categorical features which are missing
features_nan=[feature for feature in df2.columns if df2[feature].isnull().sum()>1 and df2[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df2[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label               we can add one more feature  for each to give importance bt it leads to curse of dimensionality
def replace_cat_feature(dataset,features_nan):
    data=dataset.copy()
    data[features_nan]=data[features_nan].fillna('Missing')
    return data

df2=replace_cat_feature(df2,features_nan)

df2[features_nan].isnull().sum()
```

```python
### for numeric feature
features1_nan=[feature for feature in df2.columns if df2[feature].isnull().sum()>1 and df2[feature].dtypes!='O' ]

for feature in features1_nan:
    print("{}: {}% missing values".format(feature,np.round(df2[feature].isnull().mean(),4)))
```

```python
features1_nan={'LotFrontage','MasVnrArea','BsmtFullBath','BsmtHalfBath','GarageYrBlt'}
```

```python
df2['SalePrice'].isnull().sum()
```

```python
def impute_nan(df,features1_nan):
    for variable in features1_nan:
        median=df[variable].median()
        df[variable+ '_median']=df[variable].fillna(median)
        df[variable+"_random"]=df[variable]
    ##It will have the random sample to fill the na
        random_sample=df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    ##pandas need to have same index in order to merge the dataset
        random_sample.index=df[df[variable].isnull()].index
        df.loc[df[variable].isnull(),variable+'_random']=random_sample
    return df
```

```python
df2=impute_nan(df2,features1_nan)
```

```python
outliers=[]
def detect_outliers(data):
    
    threshold=3
    mean =df2[data].mean()
    std =df2[data].std()
    
    
    for i in df2[data]:
        z_score= (i - mean)/std 
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
features1_nan={'LotFrontage_random','MasVnrArea_random','BsmtFullBath_random','BsmtHalfBath_random','GarageYrBlt_random'}
for dataset in features1_nan:
    outlier_pt=detect_outliers(dataset)
    print(outlier_pt)
```
